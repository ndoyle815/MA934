```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

**Finite difference approximations**

```python
def FDE(x, f, h, method = 'Forward'):
    
    fprime = np.zeros((len(x)))
    
    if method == 'Forward':
        fprime[:-1] = (f(x[1:]) - f(x[:-1]))/h
        return np.array([x[:-1],fprime[:-1]])
    
    if method == 'Backward':
        fprime[1:] = (f(x[1:]) - f(x[:-1]))/h
        return np.array([x[1:],fprime[1:]])
    
    if method == 'Central':
        fprime[1:-1] = (f(x[2:]) - f(x[:-2]))/(2*h)
        return np.array([x[1:-1],fprime[1:-1]])
    
    if method == '5-stencil':
        fprime[2:-2] = (- f(x[4:]) + 8*f(x[3:-1]) - 8*f(x[1:-3]) + f(x[:-4]))/(12*h)
        return np.array([x[2:-2],fprime[2:-2]])
```

```python
def f(x):
    return np.sin(5*x)

def df(x):
    return 5*np.cos(5*x)
```

```python
h = 0.0125
xs = np.arange(0,2*np.pi,h)

FD = FDE(xs, f, h, method = 'Forward')
BD = FDE(xs, f, h, method = 'Forward')
CD3 = FDE(xs, f, h, method = 'Forward')
CD5 = FDE(xs, f, h, method = 'Forward')
```

```python
# Plot results
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(12, 8))

plt.plot(FD[0,:], FD[1,:], label=r'Forward-difference')
plt.plot(BD[0,:], BD[1,:], label=r'Backward-difference')
plt.plot(CD3[0,:], CD3[1,:], label=r'Central-difference')
plt.plot(CD5[0,:], CD5[1,:], label=r'5-point stencil')
plt.plot(xs, df(xs), 'k--', label=r'$f^\prime (x)$')
#plt.plot(x_fine, home_p(x_fine), "k-", linewidth=2.0, label=r"Home-made Lagrange interpolant")
#plt.plot(x_fine, inbuilt_p(x_fine), "ks", markersize = 4, label=r"In-built Lagrange interpolant")
#plt.plot(x_fine, y_fine, "--", color = (0, 0.447, 0.741, 1), label=r"Target function")
#plt.plot(x, y, "ko", markersize = 12, markerfacecolor=(0, 0.447, 0.741, 1), label=r"Initial data")

plt.legend(loc="upper left")

plt.xlabel(r"$x$")
plt.ylabel(r"$f^\prime (x)$")
#plt.xlim([xStart-1, xEnd+1])
#plt.ylim([-1, 3])
plt.grid()
plt.show()
```

**Numerical Integration**

```python
def Integrate(f, a, b, n, method = 'Trapezoidal'):
    h = (b - a)/n
    x = np.linspace(a, b, n+1)
    
    if method == 'Trapezoidal':
        return h*(f(x[0])/2 + f(x[-1])/2 + np.sum(f(x[1:-1])))
        
    if method == 'Simpsons':
        int_sum = 0
        for i in range(1,n+1,2):
            int_sum += (2/3)*f(x[i+1]) + (4/3)*f(x[i])
        return h*(f(x[0])/3 + f(x[-1])/3 + int_sum)
```

```python
def g(x):
    return x**3 - 9*x**2 + 5*x + 6
```

```python
ns = np.arange(10,510,10)
truth = 5.75

I_TRs = [np.float32] * len(ns)
I_SIs = [np.float32] * len(ns)

err_TRs = [np.float32] * len(ns)
err_SIs = [np.float32] * len(ns)

k = 0
for j in ns:
    I_TRs[k] = Integrate(g, 0, 1, ns[k], method = 'Trapezoidal')
    I_SIs[k] = Integrate(g, 0, 1, ns[k], method = 'Simpsons')
    
    err_TRs[k] = np.abs(I_TRs[k] - truth)**2
    err_SIs[k] = np.abs(I_SIs[k] - truth)**2
    k += 1
```

```python
# Plot results
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(12, 8))

plt.plot(ns,err_TRs)

plt.plot(ns, err_TRs, "--bs", markersize = 10, markerfacecolor=(0, 0.447, 0.741, 1), label=r"Trapezoidal")
plt.plot(ns, err_SIs, "--ro", markersize = 10, markerfacecolor=(0.85, 0.325, 0.098, 1), label=r"Simpsons")

plt.legend(loc="lower left")

plt.xlabel(r"Sample points $n$")
plt.ylabel(r"$E_n$")
plt.yscale('log')
#plt.xlim([xs[0]-1, xs[-1]+1])
#plt.ylim([1e-17, 1e-6])
plt.grid()
plt.show()
```
