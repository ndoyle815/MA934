**INSERTION METHOD**

```python
def SortINSR(a):
    n = len(a)
    if n == 1:
        return a
    else:
        for j in range(1,n):
            k = j
            while a[k] < a[k-1] and k > 0:
                hold = a[k-1]
                a[k-1] = a[k]
                a[k] = hold
                k -= 1
        
        return a
```

```python
SortINSR(a)
```

**SHELL SORT**

```python
def SortSHELL(a):
    n = len(a)
    if n == 1:
        return a
    else:
        #solve for imax
        #imax = int(np.floor(np.log2(n)-1))
        #Q = np.array([2**i for i in range(imax,-1,-1)])
        imax = int(np.floor(np.log(n+1)/np.log(3)))
        Q = np.array([int(0.5*(3**i-1)) for i in range(imax,-1,-1)])
        newarr = np.zeros(n)
        for j in range(len(Q)):
            #increment
            q = Q[j]
            for k in range(q):
                newarr[k:n+1-(q-k):q] = SortINSR(a[k:n+1-(q-k):q])
        
        return newarr
```

```python
SortSHELL(a)
```

```python
#Interlacing function
def interlace(list1, list2):
    alist = []
    if (len(list1) == 0):
        return list2
    elif (len(list2) == 0):
        return list1
    elif list1[0] < list2[0]:
        alist.append(list1[0])
        return alist + interlace(list1[1:], list2)
    else:
        alist.append(list2[0])
        return alist + interlace(list1, list2[1:])
    
#merge function
def SortMERGE(a):
    n = len(a)
    if n == 1:
        return a
    else:
        m = int(n/2)
        return interlace(SortMERGE(a[0:m]),SortMERGE(a[m:n]))
```

```python
#timing
import timeit
m = 3000
sizes = np.arange(1,m+1,100)

#insertion method
runtimesINSR = [np.float32(0.0)] * int(m/100)
j=0

for i in sizes:
    A = np.random.randint(int(m*10**3), size=i)
    starttime = timeit.default_timer()
    resultINSR = SortINSR(A)
    runtimesINSR[j] = timeit.default_timer() - starttime
    j=j+1

#shell sort
runtimesSHELL = [np.float32(0.0)] * int(m/100)
j=0

for i in sizes:
    A = np.random.randint(int(m*10**3), size=i)
    starttime = timeit.default_timer()
    resultSHELL = SortSHELL(A)
    runtimesSHELL[j] = timeit.default_timer() - starttime
    j=j+1
    
#merge sort
runtimesMERGE = [np.float32(0.0)] * int(m/100)
j=0

for i in sizes:
    A = np.random.randint(int(m*10**3), size=i).tolist()
    starttime = timeit.default_timer()
    resultMERGE = SortMERGE(A)
    runtimesMERGE[j] = timeit.default_timer() - starttime
    j=j+1
```

```python
#plotting
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(12, 8))

plt.plot(sizes, [1e6 * i for i in runtimesINSR], "-ks", label=r"Computed runtime (Insertion)", markersize=10, markerfacecolor=(0, 0.447, 0.741, 1))
plt.plot(sizes, [1e6 * i for i in runtimesSHELL], "-ks", label=r"Computed runtime (Shell Sort)", markersize=10, markerfacecolor=(0.85, 0.325, 0.098, 1))
plt.plot(sizes, [1e6 * i for i in runtimesMERGE], "-ks", label=r"Computed runtime (Merge Sort)", markersize=10, markerfacecolor='green')

plt.plot(sizes, np.power(sizes, 2.0), "--b", label=r"$n^2$")
plt.plot(sizes, np.power(sizes, 1.25), "--r", label=r"$n^{1.25}$")
plt.plot(sizes, sizes*np.log2(sizes), "--g", label=r"$n \log_2(n)$")
#plt.plot(sizes, np.power(1.618, sizes), "--r", label=r"$\phi^n$")
#plt.plot(sizes, np.power(sizes, 0.5), "--b", label=r"$\sqrt{n}$")
#plt.hlines(1, 1, m, color="g", linestyle='dashed', label=r"$O(1)$")
#plt.plot(sizes, np.power(sizes, 1.0), "--g", label=r"$n$")
#plt.loglog(sizes, np.power(sizesSTD, 3.0), "--k", label=r"$n^3$")
#plt.loglog(sizesSTD, np.power(sizesSTD, np.log2(7)), "--b", label=r"$n^{2.8}$")

plt.legend(loc="lower right")

plt.xlabel(r"$n$")
plt.ylabel(r"time $(\mu s)$")
#plt.xscale("log")
plt.yscale("log")
#plt.xlim([0, 32])
#plt.ylim([1e-1, 1e9])
plt.grid()
plt.show()
```
