## Task 1

Create a notebook and implement code that reads in parameters $\alpha$ and $n$ and produces a log plot of some samples of the function $$f(x) = x^\alpha \, \log(x)$$
at values of $x$ that increase in powers of 2 from 1 to $2^n$.

```python
# Import libraries
import time
import numpy as np
from numpy import linalg as LA
import random
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
#Parameters
n = 4
```

```python
#generating array of powers from 1 to 2^n
powers = np.array([2**i for i in range(n+1)])

#defining function to plot, taking in arguments our array of powers and a
def f(x,a):
    return x**a*np.log(x)

#plotting
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(12, 8))

for a in range(5):
    plt.loglog((powers),f(powers,a), "-s", markersize = 10, label = '%d' % a)
    
plt.legend(loc="upper left")
plt.xlabel("$x$")
plt.ylabel("f(x)")
#plt.xticks(fontsize=14)
#plt.yticks(fontsize=14)
plt.grid(True)
```

## Task 2

Write a new module containing functions that compute the $n^{th}$ term, $a_n$, in the Fibonacci sequence:
 * Iteratively
 * Recursively
 * Using memoization
 
The functions should work starting from any given values of $a_1$ and $a_2$.

Measure the run-time of each of these functions over a range of values of $n$ and produce a plot to illustrate your results.

The recursive function implementation gets slow very quickly - I could only do up to $n=48$ on my laptop. Write a recursion relation for the computational complexity of the recursive version of the task. Solve it to prove that the computational complexity grows exponentially with $n$.

```python
import timeit
m = 30
```

```python
#iterative scheme
def FibITER(n):
    if n == 1 or n == 2:
        return n-1
    
    elif n > 2:
        a = np.zeros((n+1),dtype=int)
        a[1] = 0; a[2] = 1;
        for i in range(3,n+1):
            a[i] = a[i-1] + a[i-2]

        return a[-1]
        
sizes = np.arange(1,m+1)
runtimesITER = [np.float32(0.0)] * m
j=0

for i in sizes:
    starttime = timeit.default_timer()
    resultITER = FibITER(i)
    runtimesITER[j] = timeit.default_timer() - starttime
    j=j+1
```

```python
#recursive scheme
def FibRECR(n):
    if n == 1 or n == 2:
        return n-1
    
    elif n > 2:
        return FibRECR(n-1) + FibRECR(n-2)
    
runtimesRECR = [np.float32(0.0)] * m
j=0

for i in sizes:
    starttime = timeit.default_timer()
    resultRECR = FibRECR(i)
    runtimesRECR[j] = timeit.default_timer() - starttime
    j=j+1
```

```python
#memoization scheme
memo = {1:0, 2:1}

def FibMEMO(n, memo):
    if not n in memo:
        memo[n] = FibMEMO(n-1, memo) + FibMEMO(n-2, memo)
    return memo[n]

runtimesMEMO = [np.float32(0.0)] * m
j=0

for i in sizes:
    starttime = timeit.default_timer()
    resultMEMO = FibMEMO(i,memo)
    runtimesMEMO[j] = timeit.default_timer() - starttime
    j=j+1
```

```python
#plotting
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(12, 8))

plt.plot(sizes, [1e6 * i for i in runtimesITER], "-ks", label=r"Computed runtime (Iterative)", markersize=10, markerfacecolor=(0, 0.447, 0.741, 1))
plt.plot(sizes, [1e6 * i for i in runtimesRECR], "-ks", label=r"Computed runtime (Recursize)", markersize=10, markerfacecolor=(0.85, 0.325, 0.098, 1))
plt.plot(sizes, [1e6 * i for i in runtimesMEMO], "-ks", label=r"Computed runtime (Memoization)", markersize=10, markerfacecolor='green')

#plt.plot(sizes, np.power(sizes, 2.0), "--k", label=r"$n^20$")
plt.plot(sizes, np.power(1.618, sizes), "--r", label=r"$\phi^n$")
plt.plot(sizes, np.power(sizes, 0.5), "--b", label=r"$\sqrt{n}$")
plt.hlines(1, 1, m, color="g", linestyle='dashed', label=r"$O(1)$")
#plt.plot(sizes, np.power(sizes, 1.0), "--g", label=r"$n$")
#plt.loglog(sizes, np.power(sizesSTD, 3.0), "--k", label=r"$n^3$")
#plt.loglog(sizesSTD, np.power(sizesSTD, np.log2(7)), "--b", label=r"$n^{2.8}$")

plt.legend(loc="upper left")

plt.xlabel(r"$n$")
plt.ylabel(r"time $(s)$")
#plt.xscale("log")
plt.yscale("log")
plt.xlim([0, 32])
#plt.ylim([1e-1, 1e9])
plt.grid()
plt.show()
```

## Task 3

Following a similar structure (including a function - or more!) to the Notebook presented in class for matrix-matrix multiplication, add to the respective code such that it can also:
 * execute the Strassen multiplication algorithm
 * plot the runtime on the same figure as the standard matrix-matrix multiplication results and compare the two
 
Note: you may benefit from reducing the size of the problem while you do so (to avoid unnecessarily long execution while you debug).

```python
import timeit
num_power = 10
```

```python
# function that computes the standard matrix-matrix multiplication
def MMStd(A, B):
    # Detect matrix size
    n = len(A)
    nSize = (n, n)
    # Initialise result matrix
    C = np.zeros(nSize)
    
    # Loop through the matrix elements
    for i in range(n):
        for j in range(n):     
            for k in range(n): 
                # Use standard multiplication formula
                C[i][j] += A[i][k]*B[k][j]
    return C

# This gives us the list of powers of two up to a specified amount
sizesSTD = 2 ** np.arange(num_power)

# Initialise the list of runtimes with zeros
runtimesSTD = [np.float32(0.0)] * num_power

# Initialise an iterator to zero as well
j=0

for i in sizesSTD:
    A = np.random.rand(i, i)
    B = np.random.rand(i, i)
    
    starttime = timeit.default_timer()
    CStd = MMStd(A, B)
    runtimesSTD[j] = timeit.default_timer() - starttime
    j=j+1
    
# Exercise: what could go wrong with the above method? How would you improve it?
```

```python
#function that computes matrix multiplication using Strassan's algorithm
def MMStrassan(A, B):
    # Detect matrix size
    n = len(A)
    nSize = (n, n)
    # Initialise result matrix
    C = np.zeros(nSize)
    
    #divide A,B into block matrices of components of size n/2
    A11 = A[0:int(n/2),0:int(n/2)]
    A12 = A[0:int(n/2),int(n/2):n+1]
    A21 = A[int(n/2):n+1,0:int(n/2)]
    A22 = A[int(n/2):n+1,int(n/2):n+1]

    B11 = B[0:int(n/2),0:int(n/2)]
    B12 = B[0:int(n/2),int(n/2):n+1]
    B21 = B[int(n/2):n+1,0:int(n/2)]
    B22 = B[int(n/2):n+1,int(n/2):n+1]
    
    #Strassan's algorithm
  
    M1 = MMStd((A11 + A22),(B11 + B22))
    M2 = MMStd((A21 + A22),B11)
    M3 = MMStd(A11,(B12 - B22))
    M4 = MMStd(A22,(B21 - B11))
    M5 = MMStd((A11 + A12),B22)
    M6 = MMStd((A21 - A11),(B11 + B12))
    M7 = MMStd((A12 - A22),(B21 + B22))
    
    C[0:int(n/2),0:int(n/2)] = M1 + M4 - M5 + M7
    C[0:int(n/2),int(n/2):n+1] = M3 + M5
    C[int(n/2):n+1,0:int(n/2)] = M2 + M4
    C[int(n/2):n+1,int(n/2):n+1] = M1 - M2 + M3 + M6
    
    return C

# This gives us the list of powers of two up to a specified amount
sizesSTR = 2 ** np.arange(1,num_power)

# Initialise the list of runtimes with zeros
runtimesSTR = [np.float32(0.0)] * (num_power-1)

# Initialise an iterator to zero as well
j=0

for i in sizesSTR:
    A = np.random.rand(i, i)
    B = np.random.rand(i, i)
    
    starttime = timeit.default_timer()
    CStrassan = MMStrassan(A, B)
    runtimesSTR[j] = timeit.default_timer() - starttime
    j=j+1
```

```python
#plotting

plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(12, 8))

plt.loglog(sizesSTD, [1e6 * i for i in runtimesSTD], "-ko", label=r"Computed runtime (Standard)", markersize=10, markerfacecolor=(0, 0.447, 0.741, 1))
plt.loglog(sizesSTR, [1e6 * i for i in runtimesSTR], "-ks", label=r"Computed runtime (Strassan)", markersize=10, markerfacecolor=(0.85, 0.325, 0.098, 1))
plt.loglog(sizesSTD, np.power(sizesSTD, 3.0), "--r", label=r"$n^3$")
plt.loglog(sizesSTD, np.power(sizesSTD, np.log2(7)), "--b", label=r"$n^{2.8}$")

plt.legend(loc="upper left")

plt.xlabel(r"$n$")
plt.ylabel(r"time $(s)$")
plt.xlim([0.9, 700])
plt.ylim([1e-1, 1e9])
#plt.xlim([200, 700])
#plt.ylim([3e6, 1e8])
plt.grid()
plt.show()
```

```python
#sanity check
A = np.random.rand(10,10)
B = np.random.rand(10,10)
Cstd = MMStd(A,B)
Cstr = MMStrassan(A,B)
print(Cstd)
print('\n')
print(Cstr)
np.allclose(Cstd,Cstr)
```

The computational complexity of the naive divide-and-conquer algorithm for matrix multiplication satisfies the functional equation $$F(n) = 8 F(n/2) + 4 (n/2)^2$$ with $F(1)=1$.

The corresponding equation for Strassen multiplication is $$F(n) = 7 F(n/2) + 18 (n/2)^2$$ with $F(1)=1$.

Solve these recurrence relations explicitly to prove that the computational complexity of the two algorithms are $O(n^3)$ and $O(n^{\log_2(7)})$ respectively. 

It is helpful to adopt the change of variables $n=2^p$ with $a_p = F(2^p)$ to obtain linear (albeit inhomogeneous) recursion relations.
